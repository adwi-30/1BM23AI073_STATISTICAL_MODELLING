```python
# WRITE A PYTHON PROGRAM TO IMPLEMENT LINEAR REGRESSION USING LEAST SQUARE ERROR USING SOME DATASET
import numpy as np
import matplotlib.pyplot as plt
# dataset
data = {
    "shear": [177.5, 180, 230, 170.74, 150.76, 300.23, 177.98, 168.300, 155, 147.98, 176.87, 150.90, 180.23, 120, 155.23, 874.88, 122.22, 133.56, 188.67, 190.22],
    "age": [15.50, 23.76, 8, 2, 20, 84, 56, 17.78, 18.56, 16.66, 12.8, 19.9, 21.3, 25.4, 26.8, 28, 29, 26.9, 27.5, 30]
}
x = np.array(data["shear"])
y = np.array(data["age"])
n = len(x)

mean_x = np.mean(x)
mean_y = np.mean(y)
numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)
m = numerator / denominator
c = mean_y - m * mean_x

y_pred = m * x + c

lse = np.sum((y - y_pred) ** 2)

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
print(f"Least Squares Error: {lse}")

plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Shear')
plt.ylabel('Age')
plt.title('Linear Regression using Least Squares on Provided Data')
plt.legend()
plt.show()
```
